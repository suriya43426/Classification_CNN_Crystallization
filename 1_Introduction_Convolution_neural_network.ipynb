## Convolutional Layer

แผนภูมิต่อไปนี้แสดงแนวคิดพื้นฐานในการประมวลผลภาพในเลเยอร์ Convolutional แรก ภาพที่ป้อนแสดงหมายเลข 7 และสำเนาของภาพสี่ชุดแสดงอยู่ที่นี่เพื่อให้เราเห็นได้ชัดเจนยิ่งขึ้นว่าตัวกรองถูกย้ายไปยังตำแหน่งต่างๆของภาพอย่างไร สำหรับแต่ละตำแหน่งของฟิลเตอร์ผลิตภัณฑ์ดอทจะถูกคำนวณระหว่างฟิลเตอร์และพิกเซลของรูปภาพภายใต้ฟิลเตอร์ซึ่งส่งผลให้มีพิกเซลเดียวในรูปภาพที่ส่งออก ดังนั้นการย้ายฟิลเตอร์ไปทั่วทั้งภาพอินพุตจึงทำให้เกิดภาพใหม่

น้ำหนักตัวกรองสีแดงหมายความว่าตัวกรองมีปฏิกิริยาเชิงบวกต่อพิกเซลสีดำในภาพอินพุตในขณะที่พิกเซลสีน้ำเงินหมายถึงตัวกรองมีปฏิกิริยาเชิงลบต่อพิกเซลสีดำ

ในกรณีนี้ดูเหมือนว่าฟิลเตอร์จะรับรู้เส้นแนวนอนของตัวเลข 7 หลักดังที่เห็นได้จากปฏิกิริยาที่รุนแรงกับเส้นนั้นในภาพที่ส่งออก

![Convolution example](https://github.com/Hvass-Labs/TensorFlow-Tutorials/blob/master/images/02_convolution.png?raw=1)

ขนาดขั้นตอนสำหรับการย้ายตัวกรองข้ามอินพุตเรียกว่าก้าวย่าง มีการก้าวสำหรับการเลื่อนตัวกรองในแนวนอน (แกน x) และอีกก้าวหนึ่งสำหรับการเคลื่อนที่ในแนวตั้ง (แกน y)

ในซอร์สโค้ดด้านล่างการก้าวย่างถูกตั้งค่าเป็น 1 ในทั้งสองทิศทางซึ่งหมายความว่าตัวกรองจะเริ่มต้นที่มุมบนซ้ายของรูปภาพอินพุตและจะย้าย 1 พิกเซลไปทางขวาในแต่ละขั้นตอน เมื่อฟิลเตอร์ถึงจุดสิ้นสุดของรูปภาพไปทางขวาแล้วฟิลเตอร์จะถูกย้ายกลับไปทางด้านซ้ายและ 1 พิกเซลลงในรูปภาพ สิ่งนี้จะดำเนินต่อไปจนกว่าตัวกรองจะมาถึงมุมขวาล่างของภาพอินพุตและภาพเอาต์พุตทั้งหมดจะถูกสร้างขึ้น

เมื่อฟิลเตอร์มาถึงจุดสิ้นสุดของด้านขวาและด้านล่างของรูปภาพอินพุตก็สามารถบุด้วยศูนย์ (พิกเซลสีขาว) สิ่งนี้ทำให้ภาพที่ส่งออกมีขนาดเดียวกันกับภาพอินพุต

นอกจากนี้เอาต์พุตของคอนโวลูชั่นอาจถูกส่งผ่านสิ่งที่เรียกว่า Rectified Linear Unit (ReLU) ซึ่งทำให้มั่นใจได้ว่าเอาต์พุตเป็นบวกเนื่องจากค่าลบถูกตั้งค่าเป็นศูนย์ ผลลัพธ์อาจถูกสุ่มตัวอย่างลงด้วยสิ่งที่เรียกว่า max-pooling ซึ่งพิจารณาหน้าต่างขนาดเล็ก 2x2 พิกเซลและเก็บเฉพาะพิกเซลที่ใหญ่ที่สุดเท่านั้น ซึ่งจะลดความละเอียดของภาพอินพุตลงครึ่งหนึ่งเช่น ตั้งแต่ 28x28 ถึง 14x14 พิกเซล

โปรดทราบว่าเลเยอร์ convolutional ที่สองมีความซับซ้อนมากขึ้นเนื่องจากใช้ช่องสัญญาณเข้า 16 ช่อง เราต้องการตัวกรองแยกต่างหากสำหรับแต่ละช่องสัญญาณเข้าดังนั้นเราจึงต้องใช้ตัวกรอง 16 ตัวแทนที่จะใช้เพียงตัวเดียว นอกจากนี้เราต้องการช่องสัญญาณเอาต์พุต 36 ช่องจากเลเยอร์ Convolutional ที่สองดังนั้นโดยรวมแล้วเราต้องมีตัวกรอง 16 x 36 = 576 สำหรับเลเยอร์ Convolutional ที่สอง อาจเป็นเรื่องยากที่จะเข้าใจวิธีการทำงานนี้
